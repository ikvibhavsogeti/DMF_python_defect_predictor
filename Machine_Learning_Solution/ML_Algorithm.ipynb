```python
##############################################################################################################################################################
##############################################################################################################################################################
###### Python Predictor                                                                                                                                 ######
###### Version: 1.3                                                                                                                                     ######
###### In this pirticular model, Logistical regression is used for binary classification                                                                ######
###### Author: Group                                                                                                                                    ######
##############################################################################################################################################################
##############################################################################################################################################################


# Importing Data (White wine qualtity)
import pandas as pd
data = pd.read_csv('wwine.csv', delimiter=',') #storing imprted data in attributies.
data.head() # Just checking how are first few columns of data, making sure that the output values are indeed Binary. 

# manupulating the data so that if there are non numeric values, text can be loaded like numbers as well. 
import numpy as np
dataset = np.genfromtxt('wwine.csv', delimiter=",", skip_header = True)
dataset # checking again whether dataset is correctly manipulated.

# setting the prescion to upto 2 values and then rounding it off.
np.set_printoptions(precision = 2) # does not work for too wide array
np.set_printoptions(formatter = {'float': '{: 0.1f}'.format})

# Data Check
print('')
print(dataset.shape)
print('')
print(dataset[0:5])

# Spliting the data into intput and output
X = dataset[:, :-2] #Input all colums except the outcome and qualty (on which the outcome is based)
Y = dataset[:, -1] # Outcome Column

# Splititng the data again into training data and testing data.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size= 0.7)
```

```python
# Using tenserflow / keras to bulid the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(1, input_dim = len(X_train[0, :]),activation='sigmoid'))  
# instead of sigmoid, we can also use linear, for linear regression, but i think it makes more sense to use sigmoid because of the fact that the data is binarry classification data.


```

```python
# loss mattrix is binary cause we have our outputs as 1,0. 
# Choosing rms as optimiser and metrics as accuracy.
# metrics depends on the output.

model.compile(loss='binary_crossentropy', optimizer = 'rmsprop', metrics=['accuracy'])
model.fit(X_train, y_train, epochs = 512, verbose = 1)
```

```python
# checking the validation and prediction , in this irticular case, its based on accuracy
print ('True Validation Data:')
print(y_test[:10])
prediction = model.predict(X_test)
print ('Prediction:')
print(prediction[0:10].T) 
```
